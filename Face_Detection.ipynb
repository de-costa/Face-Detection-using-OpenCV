```python
import cv2
```

```python
face_clsfr = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
source = cv2.VideoCapture("film.mp4")
```

```python
while(True):
    ret,img = source.read()
    if(ret == False):
        break

    gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_clsfr.detectMultiScale(gray,1.1,5,minSize=(40,40),maxSize=(420,420))
    for face in faces:
        x = face[0]
        y = face[1]
        w = face[2]
        h = face[3]

        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
        cv2.putText(img,"FACE",(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
        
    cv2.imshow("LIVE",img)
    cv2.imshow("GRAY",gray)
    key= cv2.waitKey(5)
    if(key==27):
        break
cv2.destroyAllWindows()
source.release()
```
